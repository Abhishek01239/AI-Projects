# Resizing the image using PILLOW library

```python
from PIL import Image
```

```python
img = Image.open('labrador-retriever.jpg')
img_resized = img.resize((200,200))
img_resized
```

```python
img_resized.save('dog_image_resize.jpg')
```

```python
img_res = mpimg.imread('dog_image_resize.jpg')
```

```python
img_res_plot = plt.imshow(img_res)
```

```python
img_res.shape
```

# Convert Color Image to Grey Scale Image using OpenCV

```python
import cv2
```

```python
img = cv2.imread('labrador-retriever.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
greyscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(greyscale_image)
```

```python
greyscale_image.shape
```
